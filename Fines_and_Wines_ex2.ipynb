## Get a single dataframe that is the union of both files that only has the following columns: country, designation, points, price, state, winery.


We get just the columns that the task asks for from the resulting dataframe from the json

```python
#List with all the columns that We want
chosen_columns = ['country','designation', 'points', 'price', 'province', 'winery']

#We select those wanted columns and make a copy() so We still have the original first dataframe untouched
new_df_json = df_json.loc[: , chosen_columns].copy()
new_df_json
```

We do the same with the second dataframe

```python
#We select those wanted columns and make a copy() so We still have the original second dataframe untouched
new_df_csv = df_csv.loc[: , chosen_columns].copy()

new_df_csv
```

```python
#Finally We unify both dataframes.
df = pd.concat([new_df_json, new_df_csv])
df
```

We compare if the number of rows of the final dataframe is equal to the rows of both previous dataframes

```python
df.shape[0] == new_df_json.shape[0] + new_df_csv.shape[0]
```

## Are there duplicates in the dataframe? Delete all duplicate rows


We see if there is duplicate rows

```python
#This is the final version of this exercise

#This variable will be useful later to see if the elimination of duplicate rows is correct
n_rows_total = df.shape[0]

#Shows the duplicate rows of the file
duplicateRows = df[df.duplicated()]

#To see this more clearly We sort by the values in winery column
duplicateRows.sort_values(by="winery")
```

```python
#Deleting duplicated rows
df.drop_duplicates(inplace=True)
df
```

```python
#Is the number of rows in the dataframe with just the duplicate rows plus the number of rows without duplicates
#equal to the original dataframe number of rows?
n_rows_total == df.shape[0] + duplicateRows.shape[0]
```

## Take again the sub-set of Spanish reserve wines, in general, in Spain the wines from Rioja (a province in the north of Spain, North Spain will appear in the dataset) have great fame, therefore , it is to be expected that this province or region is the one with the best wines in this sub-set, check whether this statement is true graphically.


We obtain the mean for each province in Spain and represent the result so We can more easily view the results.

```python
mean_points_by_province =df[df.country=='Spain'].groupby('province').points.mean()

mean_points_by_province_sorted = mean_points_by_province.sort_values()
mean_points_by_province_sorted
```

```python
#We obtain both arrays to be used as variables in the barplot
names_spain_province = np.array(mean_points_by_province_sorted.keys())
points_spain_province = np.array(mean_points_by_province_sorted)

#The values are rounded.
points_spain_province = np.around(points_spain_province , decimals=2)

#Determining the size of the figure
fig, ax = plt.subplots(1, 1, figsize=(16,4))

#A barplot is created
sns.barplot(x=names_spain_province, y=points_spain_province)

#The values of each bar are set just above them
for i in range(len(points_spain_province)):
    ax.text(i, points_spain_province[i]+1, points_spain_province[i], horizontalalignment='center', size=20)

#The limits of the axes are set for a better view
ax.set(xlim=(-1,len(points_spain_province)), ylim=(0,120))
plt.show()
```

We observe that Rioja wines can be more popular but they are not the best valued in Spain.

## Graphically show the price over the score based on the province of the subset of Spanish reserve wines. Which province has the most expensive wine? Is the north of Spain still the province with the best valued reserve wine?


The dataframe showing the reserve wines in Spain is used again

```python
df_reserva
```

```python
#To show the mean of the points each province has
df_reserva.groupby("province").points.mean()
```

To create the arrray to get the column of points of the dataframe that will be used to plot:


```python
points = np.array(df_reserva.groupby("province").points.mean())
points
```

To create the arrray to get the column of price of the dataframe that will be used to plot:


```python
df_reserva.groupby("province").price.mean()
```

To create the arrray to get the column of provinces of the dataframe that will be used to plot.


```python
price=np.array(df_reserva.groupby("province").price.mean())

province = np.array(df_reserva.groupby("province").price.mean().keys())
province
```

The dataframe to plot is created:

```python
df_reserva_price_value =pd.DataFrame({"province": province, "price": price , "points" : points})
df_reserva_price_value
```

Finally We make the plotting. We plot for each region from Spain the relationship between the mean price and the mean points given to his wines.

```python
fig, ax = plt.subplots(1, 1, figsize=(10,3))

# To create the scatter plot
sns.scatterplot(data = df_reserva_price_value , x= price , y= points , hue=province, marker="p" , s = 400, legend = False)

#to make the fontsize of the contry names in the graph bigger
sns.set_context("notebook", font_scale=1.4, rc={"font.size":15})

# To show the values of the x-axis near each point
for i, val in enumerate(price):
    plt.annotate(province[i], (price[i]+1.5, points[i]))
    
#Labels
ax.set_xlabel('PRICE', fontsize=20, fontweight='bold')
ax.set_ylabel('POINTS', fontsize=20, fontweight='bold')
#To set a grid
plt.grid(color='g', linestyle='--', linewidth=0.5)
#To set a color at the background
ax.set_facecolor('lightgrey')

plt.show()
```

The result tell us that "Spain Other" region has the best relationship, quality-price.
Wines from Andalucia are sligtly better but they are far more expensive

This results are quite accurate, because the prices without a value were set as zero at the beginning was just arond 7%. So there is little bias.
